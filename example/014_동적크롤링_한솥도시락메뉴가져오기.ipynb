```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
```

```python
url ="https://www.hsd.co.kr/menu/menu_list"
driver = wb.Chrome()
driver.get(url)
driver.page_source

# [더보기] 버튼을 클릭해서 전체페이지 내용이 출력되도록 함.
for i in range(3):
    driver.find_element_by_css_selector('a.c_05').click()
    time.sleep(2)
```

```python
soup = bs(driver.page_source,'html.parser')
menu = soup.select('h4.h')
price = soup.select('div.item-price > strong')
```

```python
for i in range(len(menu)):
    print('메뉴: ' + menu[i].text)
    print('가격: ' + price[i].text)
    print('-'*40)
```

```python
import pandas as pd
```

```python
menus =[]
prices = []
for i in range(len(menu)):
    menus.append(menu[i].text.strip())
    prices.append(price[i].text)
```

```python
data = {'menu': menus, 'price': prices}
df = pd.DataFrame(data)
df.head()
```
