## Matplotlib

```python
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
#x,y축 데이터 준비
x = [ x for x in range(21)]
y1 = [ 3*x+1 for x in x ]
```

```python
y2 = [x*x+2 for x in x]
```

```python
import random
y3 = [ random.randint(10,200) for i in range(21)]
```

```python
# 꺾은선 그래프
plt.figure()
# plt.plot(x,y2)
plt.plot(x,y2, color='r', ls='--')
plt.grid()
plt.text(2.5, 300, 'y=x*x+2')
plt.xlabel("x축") #한글표기에러 처리방법
plt.title("Title")
plt.show()
```

```python
# 산점도

plt.scatter(x,y3, marker='o', label='A')
plt.scatter(x,y2, marker='x', label='B')
plt.xlabel("x축")
plt.ylabel("y축")
plt.title("title")
plt.legend()
plt.show()
```

```python
# 데이터
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

# 그래프 기본틀
plt.figure(1, figsize=(9, 3))

# 그래프 그리기
plt.subplot(131)
plt.bar(names, values)
# plt.subplot(132)
# plt.scatter(names, values)
# plt.subplot(133)
# plt.plot(names, values)
# plt.suptitle('Categorical Plotting')
plt.plot(names, values, linewidth=2.0)
plt.show()
```
