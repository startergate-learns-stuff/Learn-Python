## 정규표현식

### 1.  re 모듈
### 2.  re.compile('패턴')  => 정규식 객체 re.RegexObject
### 3.  re.RegexObject 클래스 주요 메서드
- re.match - 패턴에 매칭되는 문자열이 있는지 확인 => MatchObject
- re.search - 패턴에 처음 매칭되는 문자열 리턴 => MatchObject
- re.findall - 패턴에 매칭되는 모든 경우 리턴
- re.sub - 특정 문자열을 검색하여 대체함
- re.split - 구분자로 나누기

### 1.모듈 임포트

```python
import re
```

```python
# 정규식 객체 생성
pattern = re.compile('패턴')
```

```python
type(pattern)
```

```python
#Matches zero or more characters at the beginning of the string
#문자열의 시작부분에서 패턴에 맞는 부분인 있는지 확인, 없으면 None
pattern.match('어떤 패턴이 들어간 문자열')
```

```python
# Scan through string looking for a match, and return a corresponding match object instance
# 문자열 전체에서 패턴에 맞는 처음 부분 리턴, 없으면 None
pattern.search('어떤 패턴이 들어간 문자열')
```

```python
# Return a list of all non-overlapping matches of pattern in string
# 문자열 전체에서 패턴에 맞는 부분 전체 리스트로 리턴
pattern.findall('어떤 패턴이 들어간 패턴이 들어간 문자열')
```

```python
# Return the string obtained by replacing the leftmost non-overlapping occurrences of pattern in string 
#by the replacement repl.
# 문자열에서 패턴에 일치하는 부분을 찾아 '교체내용'으로 '바꿀횟수'만큼 교체하여 수정된 문자열 리턴
pattern.sub('어떤 패턴이 들어간 패턴','[검열됨]',2)
```

```python
#Split string by the occurrences of pattern
# 문자열을 패턴과 일치하는 부분에서 나누어서 리스트로 리턴함
pattern.split('어떤 패턴도 없는 문자열')
```

### 5. 문자 개수 판단하기

```python
# 휴대전화번호 형식 찾기
# 광주 지역의 전화번호 찾기
phone = re.compile('[0-9]{3}\-[0-9]{4}\-[0-9]{4}')
tel = re.compile('062\-[0-9]{3}\-[0-9]{4}')
text = '010-8978-8757 062-555-5555'
print(phone.findall(text))
print(tel.findall(text))
```

### 8. 그룹 사용하기
- 패턴 안에서 정규표현식을 ( )로 묶는다.
- 매치객체.group(숫자지정) 
- 0 이나 생략 : 매칭된 문자열 한꺼번에 반환

```python
m = re.search('([0-9]+) ([0-9]+)', text)
m
```

```python
m.group(1)
```

```python
m.group(2)
```

```python
m.group()
```

```python
m.group(0)
```

```python
m.groups()
```

- 그룹이름 지정 (?P<이름>정규표현식)
- 매치객체.group('그룹이름')

```python
m = re.match('(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\((?P<arg>\w+)\)', 'print(1234)')
```

```python
m.group('func')
```

```python
m.group('arg')
```

```python
def multiple10(m):
    n = int(m.group())
    return str(n*10)
```

```python
re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8')
```

### 1. 이메일 주소 형식 검사하기

```python
emails = ['python@mail.example.com', 'python+kr@example.com',              # 올바른 형식
          'python-dojang@example.co.kr', 'python_10@example.info',         # 올바른 형식
          'python.dojang@e-xample.com',                                    # 올바른 형식
          '@example.com', 'python@example', 'python@example-com']  
```

```python
pattern = re.compile('([A-Za-z0-9\-\+_\.]+)@([A-Za-z\-]+)\.([A-Za-z\.]+)')
```

```python
for email in emails:
    print(pattern.match(email))
```

### 4. 그룹으로 추출하기 

```python
s = '''
001 김갑순 여 22 002 강길동 남 21 003 성춘향 여 20
004 이길동 남 23 005 박찬수 남 24'''
```

```python
for match in re.finditer('(그룹패턴1) (그룹패턴2) (그룹패턴3) (그룹패턴4)', s):
    print(match.group(1), match.group(2), match.group(3), match.group(4))
```

```python
for match in re.finditer('(?P<number>[0-9]{3}) (?P<name>\w+) (?P<gender>남|여) (?P<age>[0-9]{2})', s): 
    print(match.group('number'),match.group('name'),match.group('gender'),match.group('age'))
```
