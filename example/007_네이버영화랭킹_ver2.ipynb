### 문제 
- 날짜별 특정 영화의 평점 데이터를 조사하여 자료를 수집하고 그래프를 그려보고자 한다.

### 1. 모듈임포트

```python
from bs4 import BeautifulSoup as bs
import requests as req
from urllib.request import urlopen
```

### 2. url 분석

https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190122
- 메인주소와 쿼리부분의 날짜로 구분할 수 있다.
- 날짜 목록이 필요하다.

```python
date_list = ['20181101', '20181107','20181114', '20181121','20181128','20181207', '20181214','20190101']
```

### 3. 날짜별 영화평점 자료를 수집

```python
m_date = []  
titles = []
points = []

for day in date_list:
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date="+day
    soup = bs(urlopen(url).read(),'html.parser')
    
    title_list = soup.select('div.tit5 > a')
    point_list = soup.findAll('td', class_ = 'point')
    
    for index in range(len(title_list)):
        titles.append(title_list[index].text.strip())
        points.append(point_list[index].text)
        m_date.append(day)
    
```

```python
m_date
```

```python
titles
```

```python
# 문자열을 숫자로 자료 변환
points = list(map(lambda x : float(x), points))
points
```

### 4. 수집한 자료 데이터프레임 생성하기

```python
import pandas as pd
```

```python
data = {'date':m_date, 'title': titles, 'point': points}
df = pd.DataFrame(data)
```
