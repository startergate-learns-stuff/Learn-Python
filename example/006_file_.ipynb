## 파일입출력

- 파일 입출력을 위해 1) 위치를 확인하고 2) 버퍼를 준비하는 과정을 open
- open(파일경로, 모드)
- 모드 r : 파일을 읽는다. 파일이 없으면 예외가 발생한다.
- 모드 w : 파일에 기록한다. 파일이 이미 있으면 덮어쓴다.
- 모드 a : 파일에 데이터를 추가한다.
- 모드 x : 파일에 기록하되 파일이 이미 있으면 실패한다.
- 파일종류지정문자 : t(텍스트 파일), b(이진 파일)
- 파일 입출력 절차 : 1) open  2) 읽기 또는 쓰기 3) close

## 파일 읽기 : readline() 
- 한 줄씩 읽으며 파일 마지막에 도달하면 빈 문자열이 리턴된다.

```python

f = open("live.txt", "r")

line = f.readline()
print(line)

f.close()
```

```python
f = open("live.txt","rt")
text = ""
line = 1
while True:
    row = f.readline()
    if not row: break
    text += str(line)+":"+row
    line += 1
f.close()
print(text)
```

```python
# readline을 사용하여 전체 데이터를 가지고 오기
f = open("mbox-short.txt", "r")

while True:
    line = f.readline()
    
    if not line:
        break
    
    print(line)

f.close()
```

```python
type(line)
```

```python
f = open("mbox-short.txt", "rt")
while True:
    text = f.read(10) #10문자씩 읽기
    if len(text) == 0: break
    print(text, end='')
f.close()
```

- 파일도 리스트와 같은 컬렉션이어서 파일 자체에 대해 루프를 돌 수 있다.

## with문
- 파일 입출력이 끝나면 반드시 close 함수로 닫아야 한다.
- 프로그램을 종료할 때 자동으로 닫지만 파일을 연속적으로 사용할 때는 꼭 닫아야 한다.
- 쓰기 모드로 기록한 후 다시 읽을 때는 일단 닫아야 읽기 모드로 다시 열 수 있다.

```python
with open("mbox-short.txt", "r") as f:
    text = f.read()
text
```

## 연습문제 1
- 다음과 같이 score.txt로 파일 저장 후에 숫자 값을 모두 읽어 총합과 평균 출력하기

```python
# 50점에서 100점 사이의 점수 10개를 저장하는 리스트 score
import random

scores = [random.randint(50, 100) for i in range(10)]
```

```python
# 공백을 기준으로 자료 score.txt 파일에 저장하기
f = open('score.txt', 'wt', encoding='UTF-8')
for score in scores:
    f.write(str(score)+' ')
f.close()
```

```python
# score.txt 파일에서 데이터 읽어 총합과 평균 구하기
f = open('score.txt')
scoreStr = f.readline().split()
scoreInt = [(lambda x:int(x))(i) for i in scoreStr]
sumall = sum(scoreInt)
avg = sumall / len(scoreInt)
f.close()
```

```python
print("총합:", sumall, "평균:",avg)
```

## 파일관리 유틸리티 - 특정 폴더에 있는 파일 전체 이름 바꾸기

```python
import os

path = "./images/"
files = os.listdir(path)
files
```

```python
files = os.listdir(path)
for file in files:
    temp = file.split('.');
    fileType = temp[1];
    temp = temp[0]
    temp = temp.split('-');
    temp = temp[1] + '-' + temp[0] + '.' + fileType
    print(temp)
    os.rename(path + file, path + temp)
```
