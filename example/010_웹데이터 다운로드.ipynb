## 표준모듈 이용

```python
import urllib.request
```

### - urlretrieve() 함수로 데이터 직접 저장

```python
#url과 저장 경로 지정하기
url = "https://uta.pw/shodou/img/28/214.png"
savename = "test.png"
```

```python
#다운로드
result=urllib.request.urlretrieve(url, savename)
result
```

```python
result2 = urllib.request.urlretrieve('http://python.org')
result2
```

```python
local_filename, headers = urllib.request.urlretrieve('http://python.org')
html = open(local_filename)
html
```

```python
urllib.request.urlretrieve('http://python.org', 'test.html')
html = open('test.html')
print(html)
```

### - urlopen() 함수로 읽어서 저장

```python
mem = urllib.request.urlopen(url).read()

with open('test2.png', mode="wb") as f:
    f.write(mem)
    print("저장완료!")
```

```python
with open('test2.png', mode="rb") as f:
    t = f.read()
    print(t)
    
```

## requests 모듈

```python
import requests
```

```python
img_url = requests.get('http://wikibook.co.kr/wikibook.png')
```

```python
with open("test.png", "wb") as f:
    f.write(img_url.content)
print("saved")
```

```python
note = requests.get('http://donote.co')
```

```python
with open("index.html", 'wb') as f:
    f.write(note.content)
print("saved")
```
