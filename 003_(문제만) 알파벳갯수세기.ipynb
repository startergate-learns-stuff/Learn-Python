## 2. 알파벳 순으로 정렬하여 출력하기

```python
song = song.upper()
sum = [0]*26
for ch in song:
    if ord(ch) >= 65 and ord(ch) <= 90:
        sum[ord(ch)-65] += 1;
```

## 3. 단어 개수 세기

```python
article1 = article.replace('\n', ' ')
article1 = article1.replace('"', '')
article1 = article1.replace('“', '')
article1 = article1.replace('”', '')
article1 = article1.replace('‘', '')
article1 = article1.replace('’', '')
article1 = article1.replace('<', ' ')
article1 = article1.replace('>', ' ')
article1 = article1.replace('(', ' ')
article1 = article1.replace(')', ' ')
article1 = article1.replace('?', ' ')
article1 = article1.replace(',', ' ')
article1 = article1.replace('.', ' ')
article1 = article1.split(' ')
sum1 = dict()
for word in article1:
    if word == '':
        continue
    if sum1.get(word):
        sum1[word] += 1
    else:
        sum1[word] = 1
```

```python
for count, word in enumerate(sum1):
    print(word, ": ", sum1[word])
```

```python
count = dict()
for word in article1:
    if word == '':
        continue
    count[word] = count.setdefault(word, 0)+1
count
```
