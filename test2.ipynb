```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import time as ti
# body > div.content-wrapper > article > table > tbody > tr
```

```python
driver = wb.Chrome()

url = 'https://namu.wiki/RecentChanges'
driver.get(url)
```

```python
soup = bs(driver.page_source, 'html.parser')
editlist = soup.select('body > div.content-wrapper > article > table > tbody > tr')
name = []
time = []
user = []
skipnext = 0
for list in editlist:
    if skipnext == 1:
        skipnext = 0
        continue
    if list.attrs:
        skipnext = 1
    namef = list.select_one('td:nth-of-type(1) > a:nth-of-type(1)')
    name.append(namef.string)
    userf = list.select_one('td:nth-of-type(2) a')
    user.append(userf.string)
    timef = list.select_one('td:nth-of-type(3) > time')
    time.append(timef.string)
```

```python
data = {'name':name, 'time': time, 'user': user}
df = pd.DataFrame(data)
```

```python
csvname = 'data/namuwiki_recent_change_'+ti.strftime('%Y-%m-%d %H_%M_%S', ti.localtime())+'.csv'
df.to_csv(csvname, encoding='utf-8')
```
